```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
sns.set()
import plotly.express as px

from sklearn.preprocessing import LabelEncoder
```

# Get Data

```python
df=pd.read_csv('/Users/etienneleconte/Downloads/Geographical Original of Music/default_features_1059_tracks.txt')
```

```python
df.columns=['X'+str(i) for i in range(1,69)]+['lat','long']
```

```python
df
```

```python
df['coord']=list(zip(df.lat,df.long))
```

```python
plt.figure(figsize=(12,8))
px.scatter(df,df.long,df.lat)
```

```python
df['SA']=np.where((df.lat.between(-58.2,11.7)) & (df.long.between(-98.1,-29.2)),'South-America','')
df['NA']=np.where((df.lat.between(30,74)) & (df.long.between(-169,-49)),'North-America','')
df['CA']=np.where((df.lat.between(10,30.1)) & (df.long.between(-125.4,-45.2)),'Central-America','')
df['AF']=np.where((df.lat.between(-36,36)) & (df.long.between(-24,54)),'Africa','')
df['EU']=np.where((df.lat.between(36,71)) & (df.long.between(-23,46)),'Europe','')
df['NAS']=np.where((df.lat.between(36,78)) & (df.long.between(46,191)),'North-Asia','')
df['SASO']=np.where((df.lat.between(-48,36)) & (df.long.between(54,191)),'South-Asia & Oceania','')
```

```python
df['Continent']=df.SA+df.NA+df.CA+df.AF+df.EU+df.NAS+df.SASO
df.head()
```

```python
df.Continent.value_counts()
```

```python
df.drop(['coord','SA','NA','CA','AF','EU','NAS','SASO'],axis=1,inplace=True)
```

```python
df.drop('loc_labels',axis=1,inplace=True)
```
